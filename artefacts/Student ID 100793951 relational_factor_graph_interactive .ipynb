```python
# ── Cell 4: Build Graph & Compute Layout ─────────────────────────────────────

G = nx.DiGraph()
for nid, meta in NODES.items():
    G.add_node(nid, **meta)
for src, dst, rel in EDGES:
    if src in NODES and dst in NODES:
        G.add_edge(src, dst, rel=rel)

print(f'Graph: {G.number_of_nodes()} nodes | {G.number_of_edges()} edges')

# ── Compute spring layout seeded with component cluster positions
np.random.seed(42)

# Initial position hints to guide spring layout by component
init_pos = {}
comp_centres = {
    'system':    ( 0.00,  0.00),
    'c1_legal':  (-0.55,  0.35),
    'c2_human':  ( 0.55, -0.35),
    'c3_risk':   ( 0.35,  0.55),
    'c4_reflect':(-0.55, -0.55),
    'c5_auto':   ( 0.00, -0.75),
}

rng = np.random.default_rng(42)
for nid, meta in NODES.items():
    cx, cy = comp_centres[meta['component']]
    init_pos[nid] = (cx + rng.uniform(-0.15, 0.15),
                     cy + rng.uniform(-0.15, 0.15))

pos = nx.spring_layout(G, pos=init_pos, fixed=None,
                        k=0.28, iterations=120, seed=42)

# Normalise to [-1, 1]
xs = np.array([p[0] for p in pos.values()])
ys = np.array([p[1] for p in pos.values()])
xs = 2 * (xs - xs.min()) / (xs.max() - xs.min()) - 1
ys = 2 * (ys - ys.min()) / (ys.max() - ys.min()) - 1
for i, nid in enumerate(pos):
    pos[nid] = (float(xs[i]), float(ys[i]))

print('Layout computed and normalised.')
```

```python
# ── Cell 6: Render Dark Responsive Interactive HTML Graph ────────────────────

GRAPH_JSON = json.dumps(graph_data)

HTML_PARTS = []

HTML_PARTS.append("""<div id="rfg-root" style="width:100%;max-width:1200px;margin:0 auto;background:#0D1117;border-radius:12px;font-family:'Segoe UI',Calibri,sans-serif;overflow:hidden;box-shadow:0 0 48px rgba(79,142,247,0.18);">""")

HTML_PARTS.append("""<div style="background:linear-gradient(135deg,#0D1B2A 0%,#1A275E 100%);padding:14px 20px;border-bottom:1px solid #1E3A5F;">
<div style="color:#E2E8F0;font-size:16px;font-weight:700;letter-spacing:0.3px;">7CS525 | Relational Factor Graph &nbsp;&mdash;&nbsp; All 5 PPT Components</div>
<div style="color:#475569;font-size:12px;margin-top:3px;">AI-Enabled Biometric Facial Recognition &nbsp;|&nbsp; UK Central Government &nbsp;&nbsp;&#x2022;&nbsp;&nbsp;<span id="rfg-count" style="color:#38BDF8;"></span></div>
</div>""")

HTML_PARTS.append("""<div style="background:#0D1B2A;padding:10px 16px;display:flex;flex-wrap:wrap;gap:6px;align-items:center;border-bottom:1px solid #1E3A5F;">
<span style="font-size:11px;font-weight:700;color:#475569;margin-right:4px;">FILTER:</span>
<button onclick="rfgFilter('all')" id="rfg-btn-all" data-color="#4F8EF7" style="padding:4px 14px;border-radius:20px;border:1.5px solid #4F8EF7;background:#4F8EF7;color:#0D1117;font-size:11px;cursor:pointer;font-weight:700;">All</button>
</div>""")

HTML_PARTS.append("""<div style="position:relative;">
<canvas id="rfg-canvas" style="width:100%;display:block;cursor:default;"></canvas>
<div id="rfg-tooltip" style="position:absolute;display:none;pointer-events:none;background:#0D1B2A;border:1px solid #1E3A5F;color:#E2E8F0;border-radius:10px;padding:12px 15px;max-width:290px;font-size:12px;line-height:1.6;box-shadow:0 8px 32px rgba(0,0,0,0.7);z-index:100;"></div>
</div>""")

HTML_PARTS.append("""<div id="rfg-legend" style="padding:10px 16px;background:#0D1B2A;border-top:1px solid #1E3A5F;display:flex;flex-wrap:wrap;gap:8px;align-items:center;">
<span style="font-size:10px;font-weight:700;color:#334155;">COMPONENTS:</span></div>""")

HTML_PARTS.append("""<div id="rfg-stats" style="background:#0D1117;padding:6px 16px;font-size:10px;color:#334155;border-top:1px solid #1E3A5F;"></div>""")

HTML_PARTS.append("""</div>""")

JS_CODE = """
<script>
(function(){
const DATA=__GRAPH_JSON__;
const nodes=DATA.nodes,edges=DATA.edges;
let activeComp='all',hoveredNode=null,selectedNode=null;
let scale=1,offsetX=0,offsetY=0,isDragging=false,dragStart=null,dragOffset=null;

const adjOut={},adjIn={};
nodes.forEach(n=>{adjOut[n.id]=[];adjIn[n.id]=[];});
edges.forEach(e=>{adjOut[e.source].push(e.target);adjIn[e.target].push(e.source);});

const root=document.getElementById('rfg-root');
const canvas=document.getElementById('rfg-canvas');
const ctx=canvas.getContext('2d');
const tip=document.getElementById('rfg-tooltip');

function cx(x){return offsetX+(x+1)/2*(canvas.width-120)*scale+60;}
function cy(y){return offsetY+(1-(y+1)/2)*(canvas.height-120)*scale+60;}
function sf(){return canvas.width/1000;}

function nodeVisible(n){return activeComp==='all'||n.comp===activeComp||n.id==='SYS';}

function resize(){
  const W=root.clientWidth;
  const H=Math.max(560,Math.round(W*0.72));
  canvas.width=W;canvas.height=H;canvas.style.height=H+'px';draw();
}

function drawPill(x,y,w,h,r,fill,stroke,glow){
  const hw=w/2,hh=h/2;
  if(glow){ctx.save();ctx.shadowColor=glow;ctx.shadowBlur=20;}
  ctx.beginPath();
  ctx.moveTo(x-hw+r,y-hh);ctx.lineTo(x+hw-r,y-hh);
  ctx.arcTo(x+hw,y-hh,x+hw,y-hh+r,r);
  ctx.lineTo(x+hw,y+hh-r);
  ctx.arcTo(x+hw,y+hh,x+hw-r,y+hh,r);
  ctx.lineTo(x-hw+r,y+hh);
  ctx.arcTo(x-hw,y+hh,x-hw,y+hh-r,r);
  ctx.lineTo(x-hw,y-hh+r);
  ctx.arcTo(x-hw,y-hh,x-hw+r,y-hh,r);
  ctx.closePath();
  ctx.fillStyle=fill;ctx.fill();
  if(stroke){ctx.strokeStyle=stroke;ctx.lineWidth=1.8;ctx.stroke();}
  if(glow)ctx.restore();
}

function drawShape(x,y,r,shape,fill,stroke,glow){
  if(glow){ctx.save();ctx.shadowColor=glow;ctx.shadowBlur=18;}
  ctx.beginPath();
  if(shape==='square'){ctx.rect(x-r,y-r,r*2,r*2);}
  else if(shape==='diamond'){ctx.moveTo(x,y-r*1.3);ctx.lineTo(x+r*1.1,y);ctx.lineTo(x,y+r*1.3);ctx.lineTo(x-r*1.1,y);ctx.closePath();}
  else if(shape==='hexagon'){for(let i=0;i<6;i++){const a=Math.PI/3*i-Math.PI/6;i===0?ctx.moveTo(x+r*Math.cos(a),y+r*Math.sin(a)):ctx.lineTo(x+r*Math.cos(a),y+r*Math.sin(a));}ctx.closePath();}
  else if(shape==='triangle'){ctx.moveTo(x,y-r*1.2);ctx.lineTo(x+r*1.1,y+r*0.75);ctx.lineTo(x-r*1.1,y+r*0.75);ctx.closePath();}
  else if(shape==='triangle_down'){ctx.moveTo(x,y+r*1.2);ctx.lineTo(x+r*1.1,y-r*0.75);ctx.lineTo(x-r*1.1,y-r*0.75);ctx.closePath();}
  else if(shape==='star'){for(let i=0;i<5;i++){const ao=Math.PI/180*(72*i-90),ai=Math.PI/180*(72*i-54);i===0?ctx.moveTo(x+r*Math.cos(ao),y+r*Math.sin(ao)):ctx.lineTo(x+r*Math.cos(ao),y+r*Math.sin(ao));ctx.lineTo(x+r*0.42*Math.cos(ai),y+r*0.42*Math.sin(ai));}ctx.closePath();}
  else{ctx.arc(x,y,r,0,Math.PI*2);}
  ctx.fillStyle=fill;ctx.fill();
  if(stroke){ctx.strokeStyle=stroke;ctx.lineWidth=1.8;ctx.stroke();}
  if(glow)ctx.restore();
}

function edgePoint(n,tx,ty){
  const nx=cx(n.x),ny=cy(n.y),s=sf();
  const dx=tx-nx,dy=ty-ny,len=Math.hypot(dx,dy);
  if(len<0.1)return{x:nx,y:ny};
  const ux=dx/len,uy=dy/len;
  let r;
  if(n.shape==='pill'){
    const hw=n.w*s/2,hh=n.h*s/2;
    const tx2=Math.abs(ux)>0.001?Math.abs(hw/ux):1e9;
    const ty2=Math.abs(uy)>0.001?Math.abs(hh/uy):1e9;
    r=Math.min(tx2,ty2)*0.92;
  } else {
    r=n.h*0.58*s;
  }
  return{x:nx+ux*r,y:ny+uy*r};
}

function arrowHead(x1,y1,x2,y2,sz,col){
  const dx=x2-x1,dy=y2-y1,len=Math.hypot(dx,dy);
  if(len<1)return;
  const ux=dx/len,uy=dy/len;
  const ax=x2-ux*sz,ay=y2-uy*sz;
  const px=-uy*sz*0.45,py=ux*sz*0.45;
  ctx.beginPath();ctx.moveTo(x2,y2);ctx.lineTo(ax+px,ay+py);ctx.lineTo(ax-px,ay-py);
  ctx.closePath();ctx.fillStyle=col;ctx.fill();
}

function drawTextInNode(n,x,y,s,scale2,isHL){
  const lines=n.lines;
  let fs,lh,yOff=0;
  if(n.shape==='pill'){
    const h=n.h*s*scale2;
    fs=Math.max(8.5,Math.min(12,h*0.285));
  } else if(n.shape==='square'){
    const r=n.h*0.5*s*scale2;
    fs=Math.max(8,Math.min(12,r*0.38));
  } else {
    const r=n.h*0.56*s*scale2;
    fs=Math.max(7.5,Math.min(11,r*0.30));
    if(n.shape==='triangle')yOff=r*0.12;
    if(n.shape==='triangle_down')yOff=-r*0.10;
  }
  lh=fs*1.28;
  ctx.font=`600 ${fs}px 'Segoe UI',Calibri,sans-serif`;
  ctx.textAlign='center';ctx.textBaseline='middle';
  const startY=y+yOff-(lines.length-1)*lh/2;
  lines.forEach((ln,li)=>{
    ctx.shadowColor='#000000';ctx.shadowBlur=6;
    ctx.fillStyle=isHL?'#FFFFFF':n.color;
    ctx.fillText(ln,x,startY+li*lh);
    ctx.shadowBlur=0;
  });
}

function draw(){
  const W=canvas.width,H=canvas.height,s=sf();
  ctx.clearRect(0,0,W,H);

  // Dark background
  ctx.fillStyle='#0D1117';ctx.fillRect(0,0,W,H);

  // Subtle dot-grid
  ctx.save();ctx.fillStyle='#161B22';
  const gs=36;
  for(let gx=gs;gx<W;gx+=gs)for(let gy=gs;gy<H;gy+=gs){ctx.beginPath();ctx.arc(gx,gy,1,0,Math.PI*2);ctx.fill();}
  ctx.restore();

  const hlNodes=new Set(),hlEdges=new Set();
  if(selectedNode){
    hlNodes.add(selectedNode);
    (adjOut[selectedNode]||[]).forEach(n=>hlNodes.add(n));
    (adjIn[selectedNode]||[]).forEach(n=>hlNodes.add(n));
    edges.forEach((e,i)=>{if(e.source===selectedNode||e.target===selectedNode)hlEdges.add(i);});
  }

  // Draw edges
  edges.forEach((e,i)=>{
    const src=nodes.find(n=>n.id===e.source),dst=nodes.find(n=>n.id===e.target);
    if(!src||!dst||!nodeVisible(src)||!nodeVisible(dst))return;
    const p1=edgePoint(src,cx(dst.x),cy(dst.y));
    const p2=edgePoint(dst,cx(src.x),cy(src.y));
    const dx=p2.x-p1.x,dy=p2.y-p1.y,len=Math.hypot(dx,dy);
    if(len<4)return;
    const isHL=selectedNode?hlEdges.has(i):true;
    const alpha=isHL?(selectedNode?0.88:0.38):0.05;
    ctx.save();ctx.globalAlpha=alpha;
    ctx.strokeStyle=isHL?e.color:'#1E2A3A';ctx.lineWidth=isHL&&selectedNode?1.8:0.9;
    ctx.beginPath();ctx.moveTo(p1.x,p1.y);ctx.lineTo(p2.x,p2.y);ctx.stroke();
    if(isHL)arrowHead(p1.x,p1.y,p2.x,p2.y,8*s,isHL?e.color:'#1E2A3A');
    ctx.restore();
  });

  // Draw nodes
  nodes.forEach(n=>{
    if(!nodeVisible(n))return;
    const x=cx(n.x),y=cy(n.y);
    const isDim=selectedNode&&!hlNodes.has(n.id);
    const isHov=n.id===hoveredNode,isSel=n.id===selectedNode;
    const sc2=isHov||isSel?1.12:1.0;
    const fillDark=n.id==='SYS'?'#0D1F3C':'#0D1B2A';
    const border=isSel?'#FBBF24':isHov?n.color:n.color+'88';
    const glow=(isHov||isSel)?n.glow:null;

    ctx.save();ctx.globalAlpha=isDim?0.16:1.0;

    if(n.shape==='pill'){
      drawPill(x,y,n.w*s*sc2,n.h*s*sc2,(n.h*s*sc2)*0.40,fillDark,border,glow);
    } else if(n.shape==='square'){
      drawShape(x,y,n.h*0.5*s*sc2,'square',fillDark,border,glow);
    } else {
      drawShape(x,y,n.h*0.56*s*sc2,n.shape,fillDark,border,glow);
    }
    drawTextInNode(n,x,y,s,sc2,isHov||isSel);
    ctx.restore();
  });
}

function hitTest(mx,my){
  const s=sf();
  for(let i=nodes.length-1;i>=0;i--){
    const n=nodes[i];if(!nodeVisible(n))continue;
    const x=cx(n.x),y=cy(n.y);
    if(n.shape==='pill'){
      if(Math.abs(mx-x)<n.w*s/2*1.15&&Math.abs(my-y)<n.h*s/2*1.15)return n.id;
    } else {
      if(Math.hypot(mx-x,my-y)<n.h*0.65*s)return n.id;
    }
  }
  return null;
}

function showTip(nid,mx,my){
  const n=nodes.find(x=>x.id===nid);if(!n)return;
  const comp=DATA.components.find(c=>c.id===n.comp);
  const conn=(adjOut[nid]||[]).length+(adjIn[nid]||[]).length;
  tip.innerHTML=
    '<div style="font-weight:700;font-size:13px;color:'+n.color+';margin-bottom:4px;">'+n.label+'</div>'+
    '<div style="color:#475569;font-size:10px;margin-bottom:7px;border-bottom:1px solid #1E3A5F;padding-bottom:5px;">'+n.category+' &#x2022; '+(comp?comp.label:'')+'</div>'+
    '<div style="font-size:11px;line-height:1.6;color:#CBD5E1;">'+n.detail+'</div>'+
    '<div style="color:#334155;font-size:10px;margin-top:7px;border-top:1px solid #1E3A5F;padding-top:4px;">&#x1F517; '+conn+' connections</div>';
  let tx=mx+16,ty=my-8;
  if(tx+300>canvas.width)tx=mx-308;
  if(ty+210>canvas.height)ty=my-220;
  tip.style.left=tx+'px';tip.style.top=ty+'px';tip.style.display='block';
}

canvas.addEventListener('mousemove',e=>{
  const rect=canvas.getBoundingClientRect();
  const sx=canvas.width/rect.width,sy=canvas.height/rect.height;
  const mx=(e.clientX-rect.left)*sx,my=(e.clientY-rect.top)*sy;
  if(isDragging){offsetX=e.clientX-dragStart.x-dragOffset.x;offsetY=e.clientY-dragStart.y-dragOffset.y;draw();return;}
  const hit=hitTest(mx,my);
  if(hit!==hoveredNode){hoveredNode=hit;canvas.style.cursor=hit?'pointer':'default';draw();}
  if(hit)showTip(hit,mx,my);else tip.style.display='none';
});
canvas.addEventListener('mouseleave',()=>{hoveredNode=null;tip.style.display='none';isDragging=false;draw();});
canvas.addEventListener('mousedown',e=>{
  const rect=canvas.getBoundingClientRect();
  const sx=canvas.width/rect.width,sy=canvas.height/rect.height;
  const mx=(e.clientX-rect.left)*sx,my=(e.clientY-rect.top)*sy;
  if(!hitTest(mx,my)){isDragging=true;dragStart={x:e.clientX,y:e.clientY};dragOffset={x:offsetX,y:offsetY};}
});
canvas.addEventListener('mouseup',e=>{
  if(!isDragging){
    const rect=canvas.getBoundingClientRect();
    const sx=canvas.width/rect.width,sy=canvas.height/rect.height;
    const mx=(e.clientX-rect.left)*sx,my=(e.clientY-rect.top)*sy;
    const hit=hitTest(mx,my);
    selectedNode=(hit&&hit!==selectedNode)?hit:null;draw();
  }
  isDragging=false;
});
canvas.addEventListener('wheel',e=>{
  e.preventDefault();scale=Math.max(0.3,Math.min(4.0,scale*(e.deltaY>0?0.91:1.10)));draw();
},{passive:false});

let lt=null,lo=null;
canvas.addEventListener('touchstart',e=>{e.preventDefault();lt={x:e.touches[0].clientX,y:e.touches[0].clientY};lo={x:offsetX,y:offsetY};},{passive:false});
canvas.addEventListener('touchmove',e=>{e.preventDefault();if(lt){offsetX=lo.x+e.touches[0].clientX-lt.x;offsetY=lo.y+e.touches[0].clientY-lt.y;draw();}},{passive:false});
canvas.addEventListener('touchend',e=>{lt=null;},{passive:false});

window.rfgFilter=function(comp){
  activeComp=comp;selectedNode=null;
  document.querySelectorAll('[id^=rfg-btn-]').forEach(b=>{
    const active=b.id==='rfg-btn-'+comp;
    const c=b.dataset.color||'#4F8EF7';
    b.style.background=active?c:'transparent';
    b.style.color=active?'#0D1117':c;
  });
  const vis=comp==='all'?nodes.length:nodes.filter(n=>n.comp===comp||n.id==='SYS').length;
  document.getElementById('rfg-stats').textContent=
    'Showing '+vis+' nodes | '+edges.length+' total relationships | Hover for details · Click to highlight · Scroll to zoom · Drag to pan';
  draw();
};

// Build filter buttons
const fb=document.querySelector('#rfg-root div:nth-child(2)');
DATA.components.forEach(c=>{
  const btn=document.createElement('button');
  btn.id='rfg-btn-'+c.id;btn.dataset.color=c.color;btn.textContent=c.label;
  btn.style.cssText='padding:4px 12px;border-radius:20px;border:1.5px solid '+c.color+';background:transparent;color:'+c.color+';font-size:11px;cursor:pointer;font-weight:600;';
  btn.onclick=()=>rfgFilter(c.id);fb.appendChild(btn);
});

// Legend
const leg=document.getElementById('rfg-legend');
DATA.components.forEach(c=>{
  const item=document.createElement('div');
  item.style.cssText='display:flex;align-items:center;gap:5px;cursor:pointer;';
  item.onclick=()=>rfgFilter(c.id);
  item.innerHTML='<span style="width:10px;height:10px;border-radius:2px;background:'+c.color+';display:inline-block;"></span><span style="font-size:10px;color:#64748B;">'+c.label+'</span>';
  leg.appendChild(item);
});

document.getElementById('rfg-count').textContent=nodes.length+' nodes \u00b7 '+edges.length+' relationships';
document.getElementById('rfg-stats').textContent=
  'All '+nodes.length+' nodes shown | '+edges.length+' relationships | Hover for details \u00b7 Click to highlight \u00b7 Scroll to zoom \u00b7 Drag to pan';

window.addEventListener('resize',resize);
resize();

})();
</script>
"""

HTML_TEMPLATE = '\n'.join(HTML_PARTS) + JS_CODE.replace('__GRAPH_JSON__', GRAPH_JSON)

display(HTML(HTML_TEMPLATE))
print('Dark interactive graph rendered.')
```

## How to Use the Interactive Graph

| Action | Result |
|---|---|
| **Hover** over a node | Tooltip: node label, category, PPT component, full description, connection count |
| **Click** a node | Highlights the node + all connected nodes and edges; click again or click canvas to deselect |
| **Filter buttons** (top bar) | Show only nodes from a specific PPT component (system node always visible) |
| **Scroll / pinch** | Zoom in and out |
| **Drag** canvas | Pan the graph |
| **Touch** (mobile) | Drag to pan; tap to select |
